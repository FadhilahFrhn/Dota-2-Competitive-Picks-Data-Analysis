```python
import pandas as pd
import numpy as np
```

```python
# Importing data
data_versatility = pd.read_csv('/content/drive/MyDrive/DA - 13/Current_Pro_meta.csv')
data_versatility.drop(columns=data_versatility.columns[0], axis=1, inplace=True)

# Counting the number of components in the 'Roles' column becomes the new column 'Versatility'
roles = list(data_versatility['Roles'])
rolelist = [a.split(',') for a in roles]
for b, a in enumerate(rolelist):
    rolelist[b]= [c.strip(' ') for c in a]
data_versatility['Roles'] = rolelist
data_versatility['Versatility'] = [len(a) for a in list(data_versatility['Roles'])]
data_versatility.sort_values('Win Rate', ascending = False).head()
```

```python
data_versatility.to_csv("/content/Generasi_GIGIH/DOTA/data_versatility.csv", index=False)
```

```python
# Break down the 'Roles' column into more specific rows
data_roles = data_versatility.explode("Roles")
grouped = data_roles.groupby("Roles")
data_roles.head()
```

```python
data_roles.to_csv("/content/Generasi_GIGIH/DOTA/data_roles.csv", index=False)
```

```python
# Calculates different statistics of specific columns in the data_roles group
roles_stats = grouped.agg({
    'Roles' : lambda x: len(x.unique()),
    'Name': lambda s: len(s.unique()),
    'Total Pro wins': 'mean',
    'Times Picked': 'mean',
    'Times Banned': 'mean',
    'Win Rate': 'mean'
})

# Assign an alias to the result of the aggregation
roles_stats = roles_stats.rename(columns={
    'Roles': 'Unique Roles',
    'Name': 'Unique heroes',
    'Total Pro wins': 'Average Total Pro wins',
    'Times Picked': 'Average Times Picked',
    'Times Banned': 'Times Banned',
    'Win Rate': 'Win Rate'
})
```
